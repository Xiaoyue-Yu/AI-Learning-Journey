````python
import os
from openai import OpenAI
from dotenv import load_dotenv
````

````python
load_dotenv()

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=os.getenv("OPENROUTER_API_KEY"),
)
````

````python
pittsburgh_text = """
Pittsburgh, located in western Pennsylvania, is famously known as the "Steel City" for its history in the steel industry and the "City of Bridges" for its 446 bridges. Situated at the confluence of the Allegheny, Monongahela, and Ohio rivers, it was historically a major industrial hub. Today, Pittsburgh has successfully transformed into a modern metropolis with a thriving economy driven by healthcare, education, robotics, and technology. It is home to renowned institutions like Carnegie Mellon University and the University of Pittsburgh. The city also boasts a rich cultural scene, including the Andy Warhol Museum, and has a deeply passionate sports culture.
"""

response = client.chat.completions.create(
  model="openrouter/free", 
  messages=[
    {
      "role": "system",
      
      "content": "You are a highly efficient text summarizer. Your task is to summarize the provided text in exactly one concise sentence."
    },
    {
      "role": "user",
      
      "content": f"Summarize the text delimited by triple backticks into one concise sentence.\n\n```{pittsburgh_text}```"
    }
  ],
  temperature=0 
)
````

````python
print("\n--- Total Response ---\n")
print(response)
print("\n-------------------\n")
print("\n--- AI Summary ---\n")
print(response.choices[0].message.content)
````
